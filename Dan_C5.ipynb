# Associative Rule Mining

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

data = pd.read_csv('data/log2.csv')

data.info()
data.sample(5)
```

```python
data.sample(5)

# Map allow to 1, deny, drop, and reset-both to 0
data['Action'] = data['Action'].map({'allow': 1, 'deny': 0, 'drop': 0, 'reset-both': 0})

df = data[['Source Port', 'Destination Port', 'NAT Source Port', 'NAT Destination Port', 'Action']].copy(deep=True)

df.sample(5)
```

```python
columns = ['Source Port', 'Destination Port', 'NAT Source Port', 'NAT Destination Port', 'Action']

df = pd.get_dummies(df, columns=columns)

df.sample(5)
```

```python
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head(10).sort_values('confidence', ascending=False)
```

```python
df['Action_0'].value_counts()
```

```python

target = '{\'Action_1\'}'

results_filter = rules[rules['consequents'].astype(str).str.contains(target, na=False)].sort_values(by='confidence', ascending=False)
results_filter.head(10)
```

```python
target = '{\'Action_0\'}'

results_filter = rules[rules['consequents'].astype(str).str.contains(target, na=False)].sort_values(by='confidence', ascending=False)
results_filter.head(10)
```
